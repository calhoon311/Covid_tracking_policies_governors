## Insights

There are a few insights noted from the charts above. There are three different measures that came from the tracking data- Number of deaths, positive cases, and hospitalized. As shown above there are line charts that show the number of these measures over the last year. When looking at the totals there is a correlation to size of population of the states and number of cases. When comparing the data with rates per 10,000 people we can compare the states more accurately. When looking at states with rates we can see that some states stand out from the rest with positive cases, deaths, and hospitializations. The 11 states that stand out are: AL, ID, IL, MA, MS, ND, NJ, NY, RI, SD, UT. When looking at these states we can compare the states governor's political party. Of the 11 states 7 of them have a republican governor and the other 4 have a democratic one. When comparing the sum of the states by poltical party by governors we can still that this a higher amount with death rates, positive cases, and hospitializations.
